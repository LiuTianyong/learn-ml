```python
from sklearn.preprocessing import MinMaxScaler

import pandas as pd
import numpy as np
```

```python
data = [
    [-1,2],
    [-0.5,6],
    [0,10],
    [1,18]
]
pd.DataFrame(data)
```

```python
# 归一化
scaler = MinMaxScaler()
scaler = scaler.fit(data)
result = scaler.transform(data)
result
```

```python
# 归一化
scaler = MinMaxScaler()
result = scaler.fit_transform(data)
result
```

```python
# 逆转 归一化结果
result_ = scaler.inverse_transform(result)
result_
```

```python
# 归一化到指定区间
scaler = MinMaxScaler(feature_range=[5,10])
result = scaler.fit_transform(data)
result
```

```python
# 连续性变量 二值化 与 分段
from sklearn.preprocessing import Binarizer
import pandas as pd
data = pd.read_csv('data/day08_Narrativedata.csv',index_col=0)

data['Age'] = data['Age'].fillna(data['Age'].mean())
X = data.iloc[:,0].values.reshape(-1,1)
X
```

```python
tranformer = Binarizer(threshold=30).fit_transform(X)
print(tranformer)
data['Age'] = tranformer
data.head(10)
```

```python
from sklearn.preprocessing import KBinsDiscretizer
import pandas as pd
data = pd.read_csv('data/day08_Narrativedata.csv',index_col=0)
data['Age'] = data['Age'].fillna(data['Age'].mean())
kbd = KBinsDiscretizer(n_bins=5,encode='onehot')
result = kbd.fit_transform(data['Age'].values.reshape(-1,1)).toarray()
result
```
